```python
# ============================================
# COMPLETE COLAB NOTEBOOK - COPY ALL CELLS
# ============================================

# ============================================
# CELL 1: INSTALL LIBRARIES (Run first)
# ============================================
!pip install easyocr paddleocr paddlepaddle-gpu opencv-python pillow matplotlib numpy

# ============================================
# CELL 2: IMPORT LIBRARIES
# ============================================
import easyocr
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

_paddleocr_imported_successfully = False
try:
    from paddleocr import PaddleOCR
    _paddleocr_imported_successfully = True
except RuntimeError as e:
    print(f"‚ö†Ô∏è Skipping paddleocr import in CELL 2 due to: {e}. This usually means it was initialized already.")
except Exception as e:
    print(f"‚ùå An unexpected error occurred during paddleocr import: {e}")

# ============================================
# CELL 3: INITIALIZE OCR ENGINES
# ============================================
print("üîÑ Initializing EasyOCR...")
reader = easyocr.Reader(['hi', 'en'], gpu=True)
print("‚úÖ EasyOCR ready!")

paddle_reader = None # Initialize paddle_reader to None

if _paddleocr_imported_successfully:
    print("\nüîÑ Initializing PaddleOCR...")
    try:
        # Check if paddle_reader is already an instance of PaddleOCR (if cell was run before)
        if 'paddle_reader' in globals() and isinstance(globals()['paddle_reader'], PaddleOCR):
            print("‚ö†Ô∏è PaddleOCR instance already exists. Reusing it.")
        else:
            paddle_reader = PaddleOCR(use_angle_cls=True, lang='hi', use_gpu=True)
            print("‚úÖ PaddleOCR ready!")
    except RuntimeError as e:
        print(f"‚ùå Error initializing PaddleOCR instance: {e}. PaddleOCR will be unavailable.")
    except Exception as e:
        print(f"‚ùå An unexpected error occurred during PaddleOCR initialization: {e}")
else:
    print("\n‚ö†Ô∏è PaddleOCR was not imported successfully earlier. PaddleOCR features will be unavailable.")

# ============================================
# CELL 4: DEFINE TEST_OCR FUNCTION (EasyOCR)
# ============================================
def test_ocr(image_path):
    """Test EasyOCR with preprocessing visualization"""
    # Read image
    img = cv2.imread(image_path)

    if img is None:
        print(f"‚ùå ERROR: Could not read image at '{image_path}'")
        print(f"Make sure the file exists and the path is correct!")
        return None

    # Preprocessing (improves accuracy on poor quality docs)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    denoised = cv2.fastNlMeansDenoising(gray, None, 10, 7, 21)
    binary = cv2.adaptiveThreshold(denoised, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY, 11, 2)

    # OCR
    print("üîç Running EasyOCR...")
    results = reader.readtext(binary, detail=1)

    # Display results
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(binary, cmap='gray')
    plt.title("Preprocessed")
    plt.axis('off')

    plt.subplot(1, 3, 3)
    img_with_boxes = img.copy()
    for (bbox, text, prob) in results:
        pts = [(int(bbox[i][0]), int(bbox[i][1])) for i in range(4)]
        cv2.polylines(img_with_boxes, [np.array(pts)], True, (0, 255, 0), 2)
    plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
    plt.title("OCR Output")
    plt.axis('off')
    plt.show()

    # Print extracted text
    print("\nüìÑ EASYOCR EXTRACTED TEXT:")
    print("="*50)
    for (bbox, text, prob) in results:
        print(f"{text} (confidence: {prob:.2f})")

    return results

# ============================================
# CELL 5: DEFINE TEST_PADDLE_OCR FUNCTION
# ============================================
def test_paddle_ocr(image_path):
    """Test PaddleOCR"""
    if paddle_reader is None:
        print("‚ùå PaddleOCR engine not initialized or available. Skipping test.")
        return None
    print("üîç Running PaddleOCR...")
    result = paddle_reader.ocr(image_path, cls=True)

    if result is None or len(result) == 0 or result[0] is None:
        print("‚ùå No text detected by PaddleOCR")
        return None

    print("\nüìÑ PADDLEOCR EXTRACTED TEXT:")
    print("="*50)
    for line in result[0]:
        text = line[1][0]
        confidence = line[1][1]
        print(f"{text} (confidence: {confidence:.2f})")

    return result

# ============================================
# CELL 6: UPLOAD YOUR IMAGE üì§
# ============================================
from google.colab import files
import os

print("üì§ Click 'Choose Files' below to upload your Hindi image(s)")
print("(Newspaper, handwritten text, document, etc.)\n")

uploaded = files.upload()

# List uploaded files
print("\n‚úÖ Uploaded files:")
for filename in uploaded.keys():
    print(f"  - {filename}")

# ============================================
# CELL 7: TEST YOUR IMAGES üß™
# ============================================

# OPTION 1: Test with first uploaded file automatically
if len(uploaded) > 0:
    # Get first uploaded filename
    first_image = list(uploaded.keys())[0]
    print(f"\nüéØ Testing with: {first_image}\n")

    # Test with EasyOCR
    print("="*60)
    print("TESTING EASYOCR")
    print("="*60)
    easy_results = test_ocr(first_image)

    print("\n" + "="*60)
    print("TESTING PADDLEOCR")
    print("="*60)
    paddle_results = test_paddle_ocr(first_image)

else:
    print("‚ö†Ô∏è No files uploaded. Please run CELL 6 first to upload images.")

# ============================================
# CELL 8: TEST SPECIFIC IMAGE (MANUAL)
# ============================================
# If you want to test a specific image by name, run this cell:

# Replace 'newspaper.jpg' with your actual filename
my_image = "newspaper.jpg"  # üëà CHANGE THIS to your filename

print(f"\nüéØ Testing specific image: {my_image}\n")

# Check if file exists
if os.path.exists(my_image):
    # Test with both OCR engines
    print("="*60)
    print("EASYOCR RESULTS")
    print("="*60)
    easy_results = test_ocr(my_image)

    print("\n" + "="*60)
    print("PADDLEOCR RESULTS")
    print("="*60)
    paddle_results = test_paddle_ocr(my_image)

else:
    print(f"‚ùå File not found: {my_image}")
    print(f"\nüìÇ Available files in current directory:")
    for f in os.listdir('.'):
        if f.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff')):
            print(f"  ‚úì {f}")

# ============================================
# CELL 9: TEST MULTIPLE IMAGES AT ONCE
# ============================================
# Run this if you uploaded multiple images

print("üîÑ Processing all uploaded images...\n")

for filename in uploaded.keys():
    print("\n" + "="*70)
    print(f"üìÑ PROCESSING: {filename}")
    print("="*70)

    try:
        # Test with EasyOCR
        print("\n[EasyOCR Results]")
        easy_results = test_ocr(filename)

        print("\n[PaddleOCR Results]")
        paddle_results = test_paddle_ocr(filename)

    except Exception as e:
        print(f"‚ùå Error processing {filename}: {e}")

    print("\n" + "-"*70 + "\n")

print("‚úÖ All images processed!")

# ============================================
# CELL 10: COMPARE OCR ENGINES
# ============================================
def compare_ocr_engines(image_path):
    """Compare EasyOCR vs PaddleOCR side by side"""
    print(f"\nüî¨ COMPARING OCR ENGINES ON: {image_path}")
    print("="*70)

    # EasyOCR
    print("\n1Ô∏è‚É£ EASYOCR:")
    print("-"*70)
    easy_results = reader.readtext(image_path)
    easy_text = ' '.join([text for (_, text, _) in easy_results])
    print(easy_text)

    # PaddleOCR
    print("\n2Ô∏è‚É£ PADDLEOCR:")
    print("-"*70)
    paddle_text = "PaddleOCR not available."
    if paddle_reader: # Check if paddle_reader is initialized
        paddle_results = paddle_reader.ocr(image_path, cls=True)
        if paddle_results and paddle_results[0]:
            paddle_text = ' '.join([line[1][0] for line in paddle_results[0]])
            print(paddle_text)
        else:
            print("No text detected by PaddleOCR.")
    else:
        print(paddle_text) # Print the "not available" message

    # Comparison
    print("\nüìä COMPARISON:")
    print("-"*70)
    print(f"EasyOCR detected: {len(easy_results)} text regions")
    if paddle_reader and paddle_results and paddle_results[0]: # Conditional check
        print(f"PaddleOCR detected: {len(paddle_results[0])} text regions")
    else:
        print("PaddleOCR detection count not available.")
    print(f"EasyOCR total chars: {len(easy_text)}")
    if paddle_reader and paddle_results and paddle_results[0]: # Conditional check
        print(f"PaddleOCR total chars: {len(paddle_text)}")
    else:
        print("PaddleOCR total chars not available.")

# Usage: Run this after uploading an image
# compare_ocr_engines("newspaper.jpg")

# ============================================
# CELL 11: SAVE RESULTS TO TEXT FILE
# ============================================
def save_ocr_results(image_path, output_filename="ocr_output.txt"):
    """Save OCR results to a text file"""
    results = reader.readtext(image_path)

    with open(output_filename, 'w', encoding='utf-8') as f:
        f.write(f"OCR Results for: {image_path}\n")
        f.write("="*50 + "\n\n")

        for (bbox, text, prob) in results:
            f.write(f"{text}\n")

    print(f"‚úÖ Results saved to: {output_filename}")

    # Download the file
    files.download(output_filename)

# Usage:
# save_ocr_results("newspaper.jpg")
```
